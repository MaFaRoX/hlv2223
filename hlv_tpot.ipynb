```python
#pip install tpot
import pandas as pd
import numpy as np
df = pd.read_csv(r'E:\BT\sim_train.csv')
```

```python
df
```

```python
df=df.replace(['itelecom','gmobile','vietnamobile','mobifone','vinaphone','viettel'],[1,2,3,4,5,6])
```

```python
df = df[df['price_vnd'] <= 10000000000]
df2 = df[df['price_vnd'] <= 10000000000]
```

```python
df['price_vnd'] = np.log(df['price_vnd'])
```

```python
x = df.drop('price_vnd', axis = 1)
y = df.price_vnd
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from tpot import TPOTRegressor
```

```python
tpot = TPOTRegressor(generations=5, population_size=30, verbosity=3, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
tpot.fit(x_train, y_train)
```

```python
tpot.export('tpot_log_sqr.py')
```

```python
tpot2 = TPOTRegressor(generations=5, population_size=30, verbosity=3, n_jobs=-1, scoring='neg_mean_absolute_error')
```

```python
tpot2.fit(x_train, y_train)
```

```python
tpot2.export('tpot_log_abs.py')
```
